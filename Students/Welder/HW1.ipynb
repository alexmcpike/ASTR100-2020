```python
#3
import math
import matplotlib.patches as mpatches
import numpy as np
from matplotlib import pyplot as plt
import matplotlib.patches as mpatches
def photon(N):
    x=np.random.normal(0,1,N)
    y=np.random.normal(0,1,N)
    x1=12*np.random.random(N)-6
    y1=12*np.random.random(N)-6
    
    plt.xlim(-6,6)
    plt.ylim(-6,6)
    orange_l = mpatches.Patch(color='orange', label='uniform')
    plt.legend(handles=[orange_l])
    plt.scatter(x,y)
    plt.scatter(x1,y1, alpha=0.1)
    plt.show()
    
    detector = np.zeros((100,100))
    def conversion(x,y, minmax=(-6,6)):
        gridvaluex = 100/(minmax[1]-minmax[0])*x+50
        gridvaluey=100/(minmax[1]-minmax[0])*y+50
        binx=int(gridvaluex)
        biny=int(gridvaluey)
        return binx, biny
    for x_, y_ in zip(x,y):
        binx, biny=conversion(x_, y_)
        detector[binx, biny]+=1
    for x_,y_ in zip(x1,y1):
        binx1, biny1 = conversion(x_, y_)
        detector[binx1, biny1] +=1
    
    plt.imshow(detector)
    
    
photon(1000)
```
